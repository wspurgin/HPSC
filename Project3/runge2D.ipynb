```python
a = loadtxt('avals.txt')
b = loadtxt('bvals.txt')
a_c = loadtxt('avals_cheb.txt')
b_c = loadtxt('bvals_cheb.txt')
f = loadtxt('runge.txt')
p10_reg = loadtxt('p10_reg.txt')
p20_reg = loadtxt('p20_reg.txt')
p10_cheb = loadtxt('p10_cheb.txt')
p20_cheb = loadtxt('p20_cheb.txt')
```

Plot of $$f(x, y)$$

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
fig = figure()
ax = fig.add_subplot(111, projection='3d')
X, Y = meshgrid(a, b)
surf = ax.plot_surface(X, Y, f, rstride=1, cstride=1,linewidth=0, cmap=cm.jet)
ax.set_xlabel('x')
ax.set_ylabel('y')
title('$f(x,y)$')
```

Plot both the interpolants with evenly spaced nodes $p_{10}reg(x,y)$ and $p_{20}reg(x, y)$

```python
fig = figure()
ax = fig.add_subplot(111, projection='3d')
X, Y = meshgrid(a, b)
surf = ax.plot_surface(X, Y, p10_reg, rstride=1, cstride=1,linewidth=0, cmap=cm.jet)
ax.set_xlabel('x')
ax.set_ylabel('y')
title('$p_{10}reg(x,y)$')
```

```python
fig = figure()
ax = fig.add_subplot(111, projection='3d')
X, Y = meshgrid(a, b)
surf = ax.plot_surface(X, Y, p20_reg, rstride=1, cstride=1,linewidth=0, cmap=cm.jet)
ax.set_xlabel('x')
ax.set_ylabel('y')
title('$p_{20}reg(x,y)$')
```

The error for the regularly spaced nodes interpolants $p_{20}reg(x,y)$ and $p_{10}reg(x,y)$

```python
e10_reg = abs(f-p10_reg)
fig = figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, e10_reg, rstride=1, cstride=1,linewidth=0, cmap=cm.jet)
ax.set_xlabel('x')
ax.set_ylabel('y')
title('$|f(x,y) - p_{10}reg(x,y)|$')
```

```python
e20_reg = abs(f-p20_reg)
fig = figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, e20_reg, rstride=1, cstride=1,linewidth=0, cmap=cm.jet)
ax.set_xlabel('x')
ax.set_ylabel('y')
title('$|f(x,y) - p_{20}reg(x,y)|$')
```

Now Ploting with the Chebyshev nodes $p_{10}cheb(x,y)$ and $p_{20}cheb(x,y)$

```python
fig = figure()
ax = fig.add_subplot(111, projection='3d')
X, Y = meshgrid(a_c, b_c)
surf = ax.plot_surface(X, Y, p10_cheb, rstride=1, cstride=1,linewidth=0, cmap=cm.jet)
ax.set_xlabel('x')
ax.set_ylabel('y')
title('$p_{10}cheb(x,y)$')
```

```python
fig = figure()
ax = fig.add_subplot(111, projection='3d')
X, Y = meshgrid(a_c, b_c)
surf = ax.plot_surface(X, Y, p20_cheb, rstride=1, cstride=1,linewidth=0, cmap=cm.jet)
ax.set_xlabel('x')
ax.set_ylabel('y')
title('$p_{20}cheb(x,y)$')
```
