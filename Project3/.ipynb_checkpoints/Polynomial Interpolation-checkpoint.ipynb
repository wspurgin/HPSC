Interestingly, over all, Newton's method was close to 10 times faster than Lagrange and never took longer than 0.16 seconds to both calculate the coefficients and evaluate the polynomial. As n and m increase, both methods take longer, yet particularly with Lagrange it increase exponentially when m is incrementing and n stays the same. 

Newton's method even with m = 10000 and n being first 10 than 50, or 5 times as many nodes, the difference in t is a magnitude of 20 times (whereas with Lagrange it is closer to 

The main reason for Legrange's higher times is the cost of evaluation the lagrange bases. While calculating the lagrange coefficients is constant time (since they are just the function values at that point which are provided), the cost of evaluating the basis is $O(n^2)$, and the cost of calculating that basis is $O(n)$. The cost for evaluating using the Newton's method is $O(n)$ while the coefficents (in general) is $O(n^2)$. In my experiment, even though the cost of calculating the coefficients for Newton's method is $O(n^3)$ which is large, Newton's method is still faster. Having found these results, it would have been better if I had found the Newton Coefficients using an alternate method to make Newton's method faster still. It goes to show how the evaluation portion of the interpolantion methods factor into the runtime far more than the calculation of the coefficients.

   Next in part 2, I applied the same concepts of interpolation in one dimension, to two dimensions. The lagrange2D.cpp (below) holds the method described in the project. 

The 2D version of lagrange interpolation is, as the project describes, the same as 1D but done with both the x, and y values. As can be seen above, instead of calculating one basis, we calculate the basis of x and basis of y, then taking their product and the product of our lagrange coefficient to get the interpolant value. In terms of algorithmic cost, recall that from part 1, the cost of evaluating the lagrange basis in 1D was $O(n^2)$. Now we are evaluating two lagrange bases in 2D, so then effictively even though the upper bound is now effectively $O(n^3)$ Since vectors x and y are the same length n, we iterate through all their elements ($n^2$) and calculate the lagrange basis (which takes n operations) for each vector x and y. Therefore the whole opearations take $n^2(2n) = 2(n^3) \Rightarrow O(n^3)$


In Part 3 of this lab, I experimented with the 'importance of nodes'. This being said I created two main routines named "runge_regular.cpp" and "runge_chebyshev.cpp" (both below) Both of these routines are identical minus the type of nodes they have. "runge_regular.cpp" has regular, evenly-spaced nodes from the interval $[-5,5]$ and "runge_chebyshev.cpp" has Chebyshev nodes from the interval $[-5,5]$ 

The purpose of experimenting with these different type and number of nodes was to show how sometimes evenly-spaced nodes (on certain function) perform very poorly. Chebyshev nodes are optimum nodes and will perform well where the evenly-spaced nodes will not. The graphs that illustrate this point can be found in runge2D.ipynb:

```python
os.system("make runge_regular.out | make runge_chebyshev.out | ./runge_regular.out | ./runge_chebyshev.out | ipython notebook --pylab inline runge2D.ipynb &")
```

Based on the graphs we can see that $p_{20}cheb(x,y)$ is the best approximation of $f(x,y)$, the next best is $p_{10}cheb(x,y)$ followed by $p_{10}reg(x,y)$ and lastly $p_{20}reg(x,y)$ The interpolations with Chebyshev nodes perform better as the number of nodes increases, whereas the regularly spaced nodes perform worse as the nodes increase (though even with just 10 nodes the regularly spaced interpolation is way off in this case)

Looking at the error for each of these seperate interpolations (as seen in the error graphs in runge2D.ipynb) it's apparent that the Chebyshev nodes work better, primarily because they are located more along the edges (closer to -5, and 5 in this case) Since evenly-spaced nodes have the most trouble on the bounds, the Chebyshev nodes are better at approximating close to the bounds.
