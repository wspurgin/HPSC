# Plotting Introduction IPython Notebook

First, we load our saved data files using the Numpy "loadtxt" routine (hit [shift]-[return] to execute the code block):

```python
x  = loadtxt('x.txt')
T1 = loadtxt('T1.txt')
T3 = loadtxt('T3.txt')
T5 = loadtxt('T5.txt')
T7 = loadtxt('T7.txt')
```

At any point during our notebook session, we may see which variables are currently allocated (and their respective sizes/types), with the "whos" command:

```python
whos
```

We'll first create a simple plot, with the 'x' array defining the horizontal axis points, and the 'T1' array defining the vertical axis points.  We can add easily add axis labels and a title, by ensuring that the text is enclosed in quotation marks (single or double).  In addition, we can display rendered LaTeX markup by enclosing the text in $-signs:

```python
plot(x,T1)
xlabel('x')
ylabel("y")
title('$T_1(x)$')
```

Let's go ahead and plot the remaining graphs in separate figure windows as well.  Note that we can concatenate commands with semicolons (as in C++), but that we do not need semicolons to end a line:

```python
plot(x,T3); xlabel('x'); ylabel('y'); title('$T_3(x)$')
```

```python
plot(x,T5); xlabel('x'); ylabel('y'); title('$T_5(x)$')
```

```python
plot(x,T7); xlabel('x'); ylabel('y'); title('$T_7(x)$')
```

Alternatively, we may plot all four functions as subplots within the same figure window:

```python
subplot(2,2,1); plot(x,T1); xlabel('x'); ylabel('y'); title('$T_1(x)$')
subplot(2,2,2); plot(x,T3); xlabel('x'); ylabel('y'); title('$T_3(x)$')
subplot(2,2,3); plot(x,T5); xlabel('x'); ylabel('y'); title('$T_5(x)$')
subplot(2,2,4); plot(x,T7); xlabel('x'); ylabel('y'); title('$T_7(x)$')
```

Instead, we can overlay each curve on the same plot by putting each pair of arrays one after another in the plot command:

```python
plot(x,T1,x,T3,x,T5,x,T7)
xlabel('x')
ylabel('y')
title('Chebyshev polynomials')
legend(('$T_1(x)$', '$T_3(x)$', '$T_5(x)$', '$T_7(x)$'))
```

Or we can create the same overlaid plot, with control over line styles and colors, through issuing multiple consecutive plot commands in the same input box

```python
plot(x, T1, 'b-',  label='$T_1(x)$')
plot(x, T3, 'r--', label='$T_3(x)$')
plot(x, T5, 'm:',  label='$T_5(x)$')
plot(x, T7, 'g-.', label='$T_7(x)$')
xlabel('x')
ylabel('y')
title('Chebyshev polynomials')
legend()
```
