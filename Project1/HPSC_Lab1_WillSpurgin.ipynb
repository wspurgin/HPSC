```python
semilogy(z, err_8, 'b-', label='$err_{8}(x)$')
semilogy(z, err_12, 'r--', label='$err_{12}(x)$')
xlabel('x')
ylabel('y')
title('Error in Taylor series Approximations')
legend()
```

Based on the graph of the relative error of the Taylor Polynomial function $p_{8}(x)$ and $p_{12}(x)$, it is apparent that the higher polynomial function provides a closer degree of accuracy. Thus $p_{12}(x)$ which has at $x=1$ a relative error of approximately $10^{-9}$, has roughly 4 more accurate digits over $p_{8}(x)$.

The upper bound on the error for $-1 < x < 1 =$ 

$\frac{f^{n+1}(\xi)}{(n+1)!}(x-a)^{n+1}$ 

$$= \frac{e^{\xi}}{9!}x^{9}$$

We need to pick a value for both $\xi$ and $x$ that will give us the upper bound of the error (the highest value for the above equation).
In this case, they are the same. $\xi$ is defined as being between the value $a$ and $x$ which means 1 is our best choice for $\xi$. Next the largest $x^{9}$ for the above equation that can be achieved in our bounds is $x = 1$ Therefore

$$\frac{e^{1}}{9!} \leq 10^{-M} \to \approx 7.5E06 \leq 10^{-M}$$

Taking the $-log$ of both sides give us $M \approx 5$ 

Thus our upper bound of error (as seen on the graph) is very close to this value. When $x = 1$ the error is just above $10^{-5}$. Acceptable though not perfect results.

To compute $e^{-13.2}$, we used two different methods: $p_{33}(-13.2)$  and $\frac{1}{p_{33}(13.2)}$. The first method returned a value that was grossly incorrect. The "true" value for $e^{-13.2}$ which was calculated through the "math.h" exp() function, is $1.8506E^{-06}$. The $p_{33}(x)$ function returned the value: $$0.788489$$ Which has a relative error: $$426071$$ which mean no digits in the caculated value are correct. Where as when using the second method, $$\frac{1}{p_{33}(13.2)}$$ The value returned was: $$1.85061E{-06}$$ which has a relative error of $3.28405E^{-06}$

```python
n = loadtxt("n.txt")
q = loadtxt("q.txt")
q_p = loadtxt("q_p.txt")

plot(n, q, "b-", label="$q(n)$")
plot(n, q_p, "r o", label="$q_{p}(n)$")
xlabel("n")
ylabel("Correct Digits")
title("Number of correct digits for $ln(x)$")
legend()
```

The functions $q(n)$ and $q_{p}(n)$ as $h$ goes to 0 (or as the value of $n$ increases) have the same behaviour. 
$$q(n) = -log_{10}(2r)$$
Where $r = \frac{\frac{f(a+h) - f(a)}{h} - f'(a)}{f'(a)}$


$$q_{p}(n) = -log_{10}(2R)$$
Where $R = c_1(h) - c_2(\frac{1}{h})$
$$h = 2^{-n}$$
When $n$ is small $h$ is larger and both absolute and relative erors will be large, but as $n$ gets larger and $h$ gets smaller (particularly around when $n \approx 22$) the error decreases greatly. Beyond this "sweet-spot" as $h$ gets smaller the error increases again due to the areas in the both $R$ where the second constant $c_2$ is multiplied by $\frac{1}{h}$
Then in $r$ where in the numerator the first value would have be divided by a decrementing $h$.

The first plot (shown above) makes it seem as if the two approximations are identical to the true value of $e^x$. However in the second graph we can see that, while the error is small in magnitude, it is still there. The second plot also shows the nature of the Taylor Polynomial approximation. As $x$ approaches the point at which the model was built $a = 0$ the number of accurate digits increases exponentially. That is the nature of Taylor Polynomials; when the value of $x$ is close to the value at which the model was evaluated at, the approximation is extremely accurate. However the further $x$ is from the value the model was evaluated, the more inaccurate the approximation becomes.

To illustrate this point, we were asked to approximate $e^{-13.2}$ with $p_{33}(-13.2)$ and $\frac{1}{p_{33}(13.2)}$.
